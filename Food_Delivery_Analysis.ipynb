# Food Delivery Data Analysis Hackathon

This notebook demonstrates loading CSV, JSON, and SQL datasets, merging them using LEFT JOINs, and performing analysis.

```python
import pandas as pd
import sqlite3

# Load datasets
orders = pd.read_csv('orders.csv')
users = pd.read_json('users.json')

# Load restaurant data from SQL
conn = sqlite3.connect('restaurants.db')
restaurants = pd.read_sql('SELECT * FROM restaurants', conn)

# Merge datasets using LEFT JOINs
final_df = orders.merge(users, on='user_id', how='left') \
                   .merge(restaurants, on='restaurant_id', how='left')

# Save final dataset
final_df.to_csv('final_food_delivery_dataset.csv', index=False)

final_df.head()
```
