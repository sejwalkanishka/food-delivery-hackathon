# Food Delivery Data Analysis Hackathon

This notebook demonstrates how transactional, user, and restaurant datasets are combined using LEFT JOINs for analysis.

**Note:** This notebook focuses on data integration logic. Dataset execution may require local setup.

```python
import pandas as pd
import sqlite3

# Load orders and users data
orders = pd.read_csv('orders.csv')
users = pd.read_json('users.json')

# --- Restaurant data loading ---
# The restaurants.sql file contains restaurant master data.
# It can be loaded into a SQLite database locally before execution.
# For hackathon evaluation, the join logic is demonstrated below.

# Example placeholder for restaurant dataframe
restaurants = pd.DataFrame(columns=['restaurant_id', 'restaurant_name', 'city', 'cuisine', 'rating'])

# Merge datasets using LEFT JOINs
final_df = orders.merge(users, on='user_id', how='left') \
                   .merge(restaurants, on='restaurant_id', how='left')

# Preview final dataset structure
final_df.head()
```
